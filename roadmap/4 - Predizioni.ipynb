```python
from zenshin import Zenshin
```

```python
model_lk = Zenshin(mode=1)      # Lasso + KMeans
model_rsp = Zenshin(mode=2)     # Ridge + Spectral Clustering
model_svragg = Zenshin(mode=7)  # LinearSVR + Agglomerative Clustering
```

# Charles De Ketelaere

```python
model_lk.predict(player_name="Charles De Ketelaere")
```

```python
model_rsp.predict(player_name="Charles De Ketelaere")
```

```python
model_svragg.predict(player_name="Charles De Ketelaere")
```
